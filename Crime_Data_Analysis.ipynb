```python
street_map = (
    df.dropna(subset=["STREET", "Lat", "Long"])
      .groupby(["Lat", "Long"])["STREET"]
      .agg(lambda x: x.value_counts().idxmax())
)


mask = df["STREET"].isna() & df["Lat"].notna() & df["Long"].notna()

df.loc[mask, "STREET"] = df.loc[mask].set_index(["Lat", "Long"]).index.map(street_map)

```

```python
df.isna().sum()
```

```python
reporting_map = (
    df.dropna(subset=["STREET", "REPORTING_AREA"])
      .groupby("STREET")["REPORTING_AREA"]
      .agg(lambda x: x.value_counts().idxmax())
)
mask = df["REPORTING_AREA"].isna() & df["STREET"].notna()


df.loc[mask, "REPORTING_AREA"] = df.loc[mask, "STREET"].map(reporting_map)

```

```python
df.isna().sum()
```

```python

district_map = (
    df.dropna(subset=["REPORTING_AREA", "DISTRICT"])
      .groupby("REPORTING_AREA")["DISTRICT"]
      .agg(lambda x: x.value_counts().idxmax())
)

mask = df["DISTRICT"].isna() & df["REPORTING_AREA"].notna()

df.loc[mask, "DISTRICT"] = df.loc[mask, "REPORTING_AREA"].map(district_map)
```

```python
df.isna().sum()
```

```python
reporting_map = (
    df.dropna(subset=["REPORTING_AREA", "Lat", "Long"])
      .groupby(["Lat", "Long"])["REPORTING_AREA"]
      .agg(lambda x: x.mode()[0])
)
mask = (
    df["REPORTING_AREA"].isna() &
    df["Lat"].notna() &
    df["Long"].notna()
)

df.loc[mask, "REPORTING_AREA"] = (
    df.loc[mask]
      .set_index(["Lat", "Long"])
      .index.map(reporting_map)
)
```

```python
df.isna().sum()
```

```python
# Filling Null Values of District Using Lat and long
```

```python
district_map = (
    df.dropna(subset=["DISTRICT", "Lat", "Long"])
      .groupby(["Lat", "Long"])["DISTRICT"]
      .agg(lambda x: x.mode()[0])
)

mask = (
    df["DISTRICT"].isna() &
    df["Lat"].notna() &
    df["Long"].notna()
)

df.loc[mask, "DISTRICT"] = (
    df.loc[mask]
      .set_index(["Lat", "Long"])
      .index.map(district_map)
)
```

```python
df.isna().sum()
```

```python
# now fill values  of Reporting Area with the help of District

reporting_map = (
    df.dropna(subset=["DISTRICT", "REPORTING_AREA"])
      .groupby("DISTRICT")["REPORTING_AREA"]
      .agg(lambda x: x.mode()[0])
)


mask = (
    df["REPORTING_AREA"].isna() &
    df["DISTRICT"].notna()
)

df.loc[mask, "REPORTING_AREA"] = df.loc[mask, "DISTRICT"].map(reporting_map)
```

```python
df.isna().sum()
```

```python
street_map = (
    df.dropna(subset=["REPORTING_AREA","STREET"])
      .groupby("REPORTING_AREA")["STREET"]
      .agg(lambda x: x.mode()[0])
)
mask = df["STREET"].isna() & df["REPORTING_AREA"].notna()

df.loc[mask, "STREET"] = df.loc[mask, "REPORTING_AREA"].map(street_map)


```

```python
df.isna().sum()
```

```python
street_coords = (
    df.dropna(subset=["STREET","Lat","Long"])
      .groupby("STREET")[["Lat","Long"]]
      .median()
)

df = df.join(street_coords, on="STREET", rsuffix="_fill")

mask = df["Lat"].isna() & df["Long"].isna()

df.loc[mask, "Lat"] = df.loc[mask, "Lat_fill"]
df.loc[mask, "Long"] = df.loc[mask, "Long_fill"]

df.drop(columns=["Lat_fill","Long_fill"], inplace=True)
```
